```python
#first rename E.coli to Ecoli.txt 
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
def getFreq(filename):
    
    #Calculate letter frequencies
    letter_counts={}
    with open(filename) as f:
        for line in f:
            for letter in line.lower():
                if letter not in letter_counts and letter.islower():
                    letter_counts[letter]=1
                elif letter.islower():    
                    letter_counts[letter]+=1
                    
    #Get relative frequencies:                
    total_letters=sum(letter_counts.values())
    freqs={}
    for key in letter_counts.keys():
        freqs.update({key : 100*letter_counts[key]/total_letters})
    
    #Order by value, descending
    freqs = OrderedDict(sorted(freqs.items(), key=lambda x: x[1],reverse=True) ) 

    
    return freqs                
```

```python

def plotvalues(filename):
    
    data=getFreq(filename)
    wikipedia={'E': 12.70, 'T': 9.06, 'A': 8.17, 'O': 7.51, 'I': 6.97, 'N': 6.75, 
               'S': 6.33, 'H': 6.09, 'R': 5.99, 'D': 4.25, 'L': 4.03, 'C': 2.78, 
               'U': 2.76, 'M': 2.41, 'W': 2.36, 'F': 2.23, 'G': 2.02, 'Y': 1.97, 'P': 1.93, 
               'B': 1.29, 'V': 0.98, 'K': 0.77, 'J': 0.15, 'X': 0.15, 'Q': 0.10, 'Z': 0.07}
        
    wikipedia =  {k.lower(): v for k, v in wikipedia.items()}    

    fig, ax = plt.subplots(1,2, figsize=(12, 5),sharey=True)
    plt.suptitle(filename,fontsize=16)
    
    ax[0].bar(data.keys(), data.values())
    ax[0].set_title('Input file letter frequencies')
    ax[0].set_ylabel('Percentages')
    ax[0].set_xlabel('Letter')
    ax[1].bar(wikipedia.keys(),wikipedia.values())
    ax[1].set_title('English alphabet frequencies')
    ax[1].set_xlabel('Letter')
    
    plt.show()
```

```python
filenames=['bible.txt','world192.txt','Ecoli.txt']
for filename in filenames: 
    plotvalues(filename)
```
